## 画像分類
ラーメン、うどん、蕎麦の分類を行う画像分類器を実行し、読み込んだ画像の画像分類を行う。

```
# ディレクトリを設定する
data_dir="/content/drive/My Drive/AI/image-classification/" # 自分のフォルダに書き換えてください
```

### 環境の構築

```
# ライブラリのインポート
from keras.models import load_model
from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input

import numpy as np
```

```
# パラメータの設定
IMAGE_SIZE = 224 # 画像サイズ
RESULT_SET = ['ラーメン','蕎麦','うどん']
```

```
# 事前に構築したモデルの読み込み
model = load_model(data_dir + "/image_classification_model.h5")

# ネットワークの確認
# model.summary()
```

### 画像を読み込み、分類結果を表示

```
image_file='img111.jpeg' # 任意の画像を指定してください


# 以降は修正不要
img = image.load_img(image_file, target_size=(IMAGE_SIZE, IMAGE_SIZE))
arr_img = preprocess_input(image.img_to_array(img))
predict = model.predict(np.stack([arr_img]))
pred_labels = np.argmax(predict, axis=-1)

display(img)
print()
print('読み込んだ画像は、「{}」の画像です'.format(RESULT_SET[pred_labels[0]]))
```
